```python
import pandas as pd

def load_data_generator(path, chunk_size=100000):
    """
    Generator to yield chunks of data, optimizing memory usage.
    """
    with pd.read_json(path, lines=True, chunksize=chunk_size) as reader:
        for chunk in reader:
            # Extract only necessary columns immediately to save RAM
            yield chunk

def load_data_generator_csv(path, chunk_size=100000):
    """
    Generator to yield chunks of data, optimizing memory usage.
    """
    with pd.read_csv(path, chunksize=chunk_size) as reader:
        for chunk in reader:
            # Extract only necessary columns immediately to save RAM
            yield chunk
movies_and_tvs_generator = load_data_generator('C:\\Tugas Raihan\\Latihan Python\\cross-domain-recommendation-system\\data\\raw\\meta_Movies_and_TV.jsonl')
cds_and_vinyl_generator = load_data_generator('C:\\Tugas Raihan\\Latihan Python\\cross-domain-recommendation-system\\data\\raw\\meta_CDs_and_Vinyl.jsonl')
```
